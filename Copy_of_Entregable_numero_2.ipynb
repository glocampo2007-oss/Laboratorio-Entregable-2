```python
student_grades = []

while True:
    print("\n--- Menú de Notas de Estudiantes ---")
    print("1. Ingresar nota de estudiante")
    print("2. Mostrar todas las notas y promedio") # Actualizado el texto
    print("3. Modificar nota de estudiante")
    print("4. Salir")

    choice = input("Seleccione una opción: ")

    if choice == '1':
        name = input("Ingrese el nombre del estudiante: ")
        while True:
            grade_input = input(f"Ingrese la nota de {name}: ")
            try:
                grade = float(grade_input)
                if 0 <= grade <= 100: # Assuming grades are between 0 and 100
                    student_grades.append({'name': name, 'grade': grade})
                    print(f"Nota de {name} ({grade}) ingresada exitosamente.")
                    break
                else:
                    print("Error: La nota debe estar entre 0 y 100.")
            except ValueError:
                print("Error: Por favor, ingrese un número válido para la nota.")
    elif choice == '2':
        if not student_grades:
            print("No hay notas ingresadas aún.")
        else:
            print("\n--- Notas de Estudiantes ---")
            total_grades = 0
            for student in student_grades:
                print(f"Nombre: {student['name']}, Nota: {student['grade']}")
                total_grades += student['grade']

            num_students = len(student_grades)
            try:
                average_grade = total_grades / num_students
                print(f"\nPromedio de notas: {average_grade:.2f}") # Muestra el promedio con 2 decimales
            except ZeroDivisionError:
                print("No se puede calcular el promedio porque no hay estudiantes ingresados.")
    elif choice == '3': # Lógica para modificar nota de estudiante
        attempts = 0
        max_attempts = 3
        student_found_and_modified = False

        while attempts < max_attempts and not student_found_and_modified:
            student_to_modify_name = input("Ingrese el nombre del estudiante cuya nota desea modificar: ")
            target_student = None
            for student in student_grades:
                if student['name'].lower() == student_to_modify_name.lower():
                    target_student = student
                    break

            if target_student:
                print(f"Estudiante '{target_student['name']}' encontrado. Nota actual: {target_student['grade']}")
                while True: # Loop para la entrada de la nueva nota
                    new_grade_input = input(f"Ingrese la nueva nota para {target_student['name']}: ")
                    try:
                        new_grade = float(new_grade_input)
                        if 0 <= new_grade <= 100:
                            target_student['grade'] = new_grade
                            print(f"Nota de {target_student['name']} actualizada a {new_grade}.")
                            student_found_and_modified = True
                            break # Sale del bucle de entrada de nota
                        else:
                            print("Error: La nota debe estar entre 0 y 100.")
                    except ValueError:
                        print("Error: Por favor, ingrese un número válido para la nota.")
            else:
                attempts += 1
                if attempts < max_attempts:
                    print(f"Estudiante '{student_to_modify_name}' no encontrado. Le quedan {max_attempts - attempts} intentos para encontrar al estudiante.")
                else:
                    print(f"Estudiante '{student_to_modify_name}' no encontrado. Ha agotado sus {max_attempts} intentos para modificar una nota.")
    elif choice == '4':
        print("Saliendo del programa. ¡Hasta luego!")
        break
    else:
        print("Opción no válida. Por favor, intente de nuevo.")
```
